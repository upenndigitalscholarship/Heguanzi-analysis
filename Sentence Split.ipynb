```python
import csv
from google.colab import files

input_file_path = '/content/drive/MyDrive/heguanzitext.txt'
output_file_path = '/content/drive/MyDrive/sentences.csv'

# Read the contents of the TXT file
with open(input_file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# Split the text into sentences using the Chinese full stop "。"
sentences = text.split('。')

# Remove any empty sentences
sentences = [sentence.strip() for sentence in sentences if sentence.strip()]

# Write the sentences to a CSV file
with open(output_file_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows([[sentence] for sentence in sentences])

print("Sentences have been extracted and saved to 'sentences.csv'.")

# Download the CSV file
files.download(output_file_path)
```
